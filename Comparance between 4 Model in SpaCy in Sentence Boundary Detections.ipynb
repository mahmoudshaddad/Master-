# Clearing text from noise before tokenization

```python
import re


# Sentences to delete
sentences_to_delete = [
    "Technische Universität Darmstadt",
    "Printed copies are uncontrolled",
    "— Leerseite —",

]

# Create a regular expression pattern to match the sentences
pattern = "|".join(map(re.escape, sentences_to_delete))

# Use re.sub to remove the matched sentences
new_text = re.sub(pattern, "", rawtext)

# print(new_text)
```

```python
print("The length of the Text without the cleaning : " + str(len(rawtext)) + "\n" +
      "The length of the Text with the cleaning of Tu Darmstadt & white space: " + str(len(new_text)) + "\n" +
      "The difference between the Text is: " + str(len(rawtext) - len(new_text)))
```

```python
new_text2 = re.sub(r'\s+', ' ', new_text)
```

```python
print("The length of the Text without the cleaning : " + str(len(new_text)) + "\n" +
      "The length of the Text with the cleaning of Tu Darmstadt & white space: " + str(len(new_text2)) + "\n" +
      "The difference between the Text is: " + str(len(new_text) - len(new_text2)))
```

```python
def replace_unknown_characters(text, replacement=" "):
    return text.replace('�', replacement)

# Example usage

new_text3 = replace_unknown_characters(new_text2)
print(new_text3)
```

```python
print("The length of the Text without the cleaning : " + str(len(new_text2)) + "\n" +
      "The length of the Text with the cleaning of Tu Darmstadt & white space: " + str(len(new_text3)) + "\n" +
      "The difference between the Text is: " + str(len(new_text2) - len(new_text3)))
```

```python
# Define a regular expression pattern to match the date and time format
pattern = r'\d{2}/\d{2}/\d{4} \d{2}:\d{2}:\d{2}'

# Use re.sub to replace all occurrences of the pattern with an empty string
new_text4 = re.sub(pattern, '', new_text3)

# Print the cleaned text
print(new_text4)
```

```python
print("The length of the Text without the cleaning : " + str(len(new_text3)) + "\n" +
      "The length of the Text with the cleaning of Tu Darmstadt & white space: " + str(len(new_text4)) + "\n" +
      "The difference between the Text is: " + str(len(new_text3) - len(new_text4)))
```

```python
import spacy
nlp_trf = spacy.load("de_dep_news_trf")
nlp_md = spacy.load("de_core_news_md")
nlp_sm = spacy.load("de_core_news_sm")
nlp_lg=  spacy.load("de_core_news_lg")
```

```python
# Create lists to store the sentences detected by each model
sentences_list_nlp_trf = []
sentences_list_nlp_lg = []
sentences_list_nlp_md = []
sentences_list_nlp_sm = []

models = [nlp_trf, nlp_lg, nlp_md, nlp_sm]
model_names = ["de_dep_news_trf", "de_core_news_lg", "de_core_news_md", "de_core_news_sm"]

for model, model_name in zip(models, model_names):
    doc = model(text)
    sentences = [sent.text for sent in doc.sents]
    print(f"Sentences detected using {model_name}:\n")
    for i, sentence in enumerate(sentences, start=1):
        print(f"Sentence {i}: {sentence}\n")
    
    if model_name == "de_dep_news_trf":
        sentences_list_nlp_trf = sentences
    elif model_name == "de_core_news_lg":
        sentences_list_nlp_lg = sentences
    elif model_name == "de_core_news_md":
        sentences_list_nlp_md = sentences
    elif model_name == "de_core_news_sm":
        sentences_list_nlp_sm = sentences



```

```python
print("the Number of Sentence is not a used key for  evluate the acuuracy of the model   "+ "\n" +
"The number of the sentence in the model: de_dep_news_trf : " + str(len(sentences_list_nlp_trf)) + "\n" + 
"The number of the sentence in the model: de_core_news_lg : " + str(len(sentences_list_nlp_lg)) + "\n" +
"The number of the sentence in the model: de_core_news_md : " + str(len(sentences_list_nlp_md)) + "\n" +
"The number of the sentence in the model: de_core_news_sm : " + str(len(sentences_list_nlp_sm)) + "\n" 
     )
```

why the differnt hte numebr bertween the tocke why they different and how 

```python
import pandas as pd
import numpy as np

# Find the maximum length among the lists
max_length = max(len(sentences_list_nlp_trf), len(sentences_list_nlp_lg), len(sentences_list_nlp_md), len(sentences_list_nlp_sm))

# Fill shorter lists with NaN to make them the same length
if len(sentences_list_nlp_trf) < max_length:
    sentences_list_nlp_trf += [np.nan] * (max_length - len(sentences_list_nlp_trf))
if len(sentences_list_nlp_lg) < max_length:
    sentences_list_nlp_lg += [np.nan] * (max_length - len(sentences_list_nlp_lg))
if len(sentences_list_nlp_md) < max_length:
    sentences_list_nlp_md += [np.nan] * (max_length - len(sentences_list_nlp_md))
if len(sentences_list_nlp_sm) < max_length:
    sentences_list_nlp_sm += [np.nan] * (max_length - len(sentences_list_nlp_sm))

df = pd.DataFrame({
    "trf ": sentences_list_nlp_trf,
    "lg": sentences_list_nlp_lg,
    "md": sentences_list_nlp_md,
    "sm" : sentences_list_nlp_sm
})
```
